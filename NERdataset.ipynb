```python
import utils.NERutils as NU
from transformers import AutoTokenizer

bert_model_name = "bert-base-multilingual-cased"
bert_tokenizer = AutoTokenizer.from_pretrained(bert_model_name)

train = NU.readDataset("data/train.parquet")

tags, index2tag, tag2index = NU.getVocabFeatures(train)

encodings = NU.encodeDataFrame(train, bert_tokenizer, tag2index=tag2index)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

class NERdataset(Dataset):
    def __init__(self, dataset_path: str, tokenizer: AutoTokenizer) -> None:
        self.__df = NU.readDataset(dataset_path)
        self.tags, self.index2tag, self.tag2index = NU.getVocabFeatures(self.__df)
        self.MAX_LENGTH = max(NU.findMaxLength(self.__df, tokenizer), 512)
        self.encodings = NU.encodeDataFrame(self.__df, tokenizer, self.tag2index, self.MAX_LENGTH)

    def __len__(self) -> int:
        return len(self.__df)
    
    def __getitem__(self, index: int) -> dict[torch.Tensor, torch.Tensor, torch.Tensor]:
        item = {key: torch.tensor(val[index]) for key, val in self.encodings.items()}
        return item
```

```python
trainNER = NERdataset("data/train.parquet", bert_tokenizer)
```

```python
train_params = {'batch_size': 8,
                'shuffle': True,
                'num_workers': 0,
                'pin_memory': True,
                }

training_loader = torch.utils.data.DataLoader(trainNER, **train_params)
```

```python
for batch_idx, batch in enumerate(training_loader):
    print(batch["attention_mask"])
    break
```
